```python
details = {}
noiucn = {}
for name in nonexisting:
    if type(name) == str:
        print(name)
        token = '2103c15b0319ba05e7de13619aa6b7afaa68037ab347c19591a3854fb3380a5f'
        url = 'http://apiv3.iucnredlist.org/api/v3/species/'+name+'?token='+token
        request = requests.get(url)
        if request.status_code == 200:
            response = json.loads(request.text)
            if 'result' in response:
                if(len(response['result'])>0):
                    index = 0
                    if len(response['result']) > 1:
                        for i in response['result']:
                            print(i)
                        index  = int(input("Which one?"))
                    details[name] = response['result'][index]
                else:
                    noiucn[name] = response
            else:
                print({'error':request.status_code})
```
